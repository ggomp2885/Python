```python
"""
Example solution #5 ----
ohh, a DAG, is a "directed acyclic graph"
aka, a DAG, is a graph, that goes in one direction, that doesnt have any cycles,
lol nice...

"Given a directed acyclic graph,
with n vertices numbered from 0 to n-1,
and an array edges where edges[i] = [x, y]
represents a directed edge from node x to node y

Find the smallest set of vertices from
which all nodes in the graph are reachable."

This description means:
1. a graph, 
2. given as an adjacency list (perf)
3. Find the nodes that have an indegree of 0
"""

class Solution:
    def findSmallestSetOfVertices(self, n: int, edges: List[List[int]]) -> List[int]:
        indegree = [0] * n
        for _, y in edges:
            indegree[y] += 1
        
        return [node for node in range(n) if indegree[node] == 0]

"""
Interesting, so this one literally just looks at the "2nd" value given in "edges"
And adds a +1 to the index in a secondary list, when this number is found,
Then returns a list of the indices of this list that == 0

Pretty dang simple, no dfs needed, very interesting.
"""
```

```python
"""
Okay so for this one, its basically asking, are these 2 nodes in the same connected component?
So, Im pretty sure I basically run a dfs on the "source" node, and each iteration put each node into
A "seen" set, after all nodes have been seen,

I just check if "destination" is in "seen"
and return true/false
"""
import collections

class Solution:
    def validPath(self, n: int, edges: List[List[int]], source: int, destination: int) -> bool:
        graph = collections.defaultdict(list)
        for a, b in edges:
            graph[a].append(b)
            graph[b].append(a)
        
        print(graph)

        seen = set()
        seen.add(source)

        def dfs(curr_node):
            for a_node in graph[curr_node]:
                if a_node not in seen:
                    seen.add(a_node)
                    dfs(a_node)
                    
                #seen = (1)
                #dfs(1)
                #seen = (1,0)
                #dfs(0)
                # the 1 is already there
                #seen = (1,0,2)
                #dfs(2)
                # the 2 is already there
                # the 0 is already there
                # end end end (done)
            
        dfs(source)
        print(seen)
            
        if destination in seen:
            return True
            
        return False


"""The official solution is below"""    
class Solution:
    def validPath(self, n: int, edges: List[List[int]], source: int, destination: int) -> bool:
        graph = collections.defaultdict(list)
        for a, b in edges:
            graph[a].append(b)
            graph[b].append(a)
            
        seen = [False] * n
        
        def dfs(curr_node):
            if curr_node == destination:
                return True
            if not seen[curr_node]:
                seen[curr_node] = True
                for next_node in graph[curr_node]:
                    if dfs(next_node):
                        return True
            return False
            
        return dfs(source)

"""
Alright so this is just a different way to write almost the same thing,
now technically I think this official solution is better,
because it doesnt need to iterate through 100,000 nodes
if it finds destination node after 10 nodes, its done.

So actually it does make alot of sense to write it like this,
So moving forward, I can remember that in a case like this,
whenever I am looking for the presence of a specific node,
I can just write an if statement to check for this,
at the begining of every dfs iteration,
Its going to end up being faster 

Also, the "seen = [False] * n"
Is something I 100% want to keep in mind,
with 1,000,000 nodes, this makes a huge difference.
"""
```
