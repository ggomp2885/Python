## Seaborn Plot Mastery

### Imports

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(context='poster',  # default #for quick use # sets font scale, lines width, and axis numbers, to a small/med/large/xlarge level, can be a little unpredictable
              font ='georgia',
              font_scale=1, 
              style='darkgrid',    # default
              palette='deep',
              rc={"lines.linewidth" : 1})

# this must be set or else the plots dont work, this is the only "set_" function
# I need to use, because it holds all of the relevant functions
# (sns.set_style('darkgrid') and sns.set_context('talk') are redundant functions for
# setting only one of these parameters)

# ahh, so I see the reason why they have specific set_functions,
# its for when you want to keep all the other settings the same,
# but just change 1 at time ---- so I can have an initial setup command that sets the long term things, and then a small setup that I c&p into a cell for custom manipulations
# but in order to do this, this small C&P setup would have to use multiple of the smaller "set_()" functions

# also, all of these settings statements have to be run above the sns.plot() function, 
# they have to be set before the "display" command
```

## Colors 

in seaborn, the "hue" argument refers to using different colors for different categories of data, "hue"=df['column'], creates different colors for each descrete catagory in this column - this is what helps in building "double/triple plots"

## Loading example dataset

```python
#sns.get_dataset_names()    #shows example datasets available from seaborn

df = sns.load_dataset('titanic')
display(df.head(), df.shape, type(df))
```

## Mock Plot to show size / style changing

run above cell, first, to load in a dataframe

```python
# countplot
sns.set(context='talk',
        font ='georgia',
        font_scale=1.1, 
        style='darkgrid',
        palette='dark')


# sns.despine

plt.figure(figsize=(2,2))        # this has to be in the same cell as the plot being made for it to work

print(f' countplot (counts number of rows with each entry) {sns.countplot(x="sex", data=df)}')
```

## Plot examples

```python
plt.figure(figsize=(3,3))


# Countplot
print(f' countplot (counts number of rows with each entry) {sns.countplot(x="sex", data=df)}')
# the y is left out, because countplot finds the unique entries in the column
# to use as the x values, and does a count of each entry to use as their respective y values
```

```python
plt.figure(figsize=(3,3))

# Barplot
print(f'Barplot - for comparing  {sns.barplot(x="sex", y="survived", hue="pclass", data=df)}')

# okay so with these you can summarize averages of each column with continuos y values too
# print(f' Barplot - for comparing  {sns.barplot(x="sex", y="fare", data=df)}')
```

## Correlation plot

```python
plt.figure(figsize=(8,8))
sns.set(context = 'paper')

print(f'Barplot - for comparing  {sns.heatmap(df.corr(), annot=True, cmap="twilight_shifted")}')
#df.corr() automatically drops any columns that arent numeric
#cmap="RdYlGn"
```

## Line Plots

```python
import pandas as pd
plt.figure(figsize=(4,4))
sns.set(context = 'paper')

df = pd.DataFrame({'train_error': [100,80,50,32,22,15,10,7,5,5,5],
                   'validation_error': [100,85,60,40,30,25,22,15,14,13,12],
                   'samples' : [0,1000,2000,3000,4000,5000,6000,7000,8000,9000,10000]})
sns.lineplot(x ='train_samples', y=['train_error','val_error'], data=df)
# okay so it looks like sklearn has a builtin function for this
```
