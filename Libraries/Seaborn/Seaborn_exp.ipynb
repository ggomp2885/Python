## Seaborn Settings & Custom functions

```python
# Seaborn Styling settings
sns.set(context='poster',          # default     # sets font scale, lines width, and axis numbers, to a small/med/large/xlarge level, can be a little unpredictable
              font ='arial',
              font_scale=1, 
              style='darkgrid',    # default
              palette='deep',
              rc={"lines.linewidth" : 1})


# custom function to put numeric labels on the bars
# argument here looks like df['sex']
def labels_on_bars(df_column_name, ax_or_plt):
    for i in range(df_column_name.nunique()):
        ax_or_plt.text(i, 
                 df_column_name.value_counts()[i],
                 df_column_name.value_counts()[i], 
                 ha='center', va='bottom', size=16)
    return
```

## Loading example dataset

```python
#sns.get_dataset_names()    #shows example datasets available from seaborn

df = sns.load_dataset('titanic')
display(df.head(), df.shape, type(df))
```

## Mock Plot to show size / style changing

run above cell, first, to load in a dataframe

```python
# each plot setup code
plt.figure(figsize=(3,3))        # this has to be in the same cell as the plot being made for it to work
sns.set(context='talk',          # this will stay the same as previous plot, until set() is called again
        font_scale=1.1,
        font='arial',
        palette='dark',
        rc={"lines.linewidth" : 1})


# sns.despine                   # removes the lines at the edges of the plots

ax = sns.countplot(x="sex", data=df)

plt.ylim(0, 700)                # custom y limit
labels_on_bars(df['sex'])
```

## Plot examples
####  Countplots and histplots are basically the same thing
a countplot is pre-setup to work better with discrete numbers,

a histplot is pre-setup to work better with continous values

## Countplot

```python
plt.figure(figsize=(3,3))

sns.countplot(x="sex", data=df)

plt.ylim(0, 700)
labels_on_bars(df['sex'])

# the y is left out, because countplot finds the unique entries in the column
# to use as the x values, and does a count of each entry to use as their respective y values
```

## Barplot

```python
plt.figure(figsize=(3,3))

sns.barplot(x="sex", y="survived", hue="pclass", data=df)

# okay so with these you can summarize averages of each column with continuos y values too
# print(f' Barplot - for comparing  {sns.barplot(x="sex", y="fare", data=df)}')
```

## Line Plots

```python
import pandas as pd
plt.figure(figsize=(4,4))
sns.set(context='talk',
        font_scale=1.1,
        font='georgia',
        palette='dark',
        rc={"lines.linewidth" : 1})


df = pd.DataFrame({'train_error': [100,80,50,32,22,15,10,7,5,5,5],
                   'validation_error': [100,85,60,40,30,25,22,15,14,13,12],
                   'examples' : [0,1000,2000,3000,4000,5000,6000,7000,8000,9000,10000]})

display(df.head(2))
print(sns.lineplot(x ='examples', y='train_error', data=df))
# okay so it looks like sklearn has a builti n function for this
```

## Multi-Plots

```python
sns.set(context='notebook',
        font_scale=1.4,
        font='georgia',
        palette='dark',
        rc={"lines.linewidth" : 1})



fig, ax = plt.subplots(2,1, figsize=(7,15))         # figsize is now located here
#countplot
ax[0].set_title('Count of male/female')
sns.countplot(x="sex", data=df, ax=ax[0])
ax[0].set_xlabel('')

labels_on_bars(df['sex'], ax[0])

#barplot
ax[1].set_title('Survival rate, by sex and pclass')
sns.barplot(x="sex", y="survived", hue="pclass", data=df, ax=ax[1])
ax[1].set_xlabel('')

# labels_on_bars(df['survived'], ax[1])
for p in ax[1].patches:
    ax[1].annotate(round(p.get_height(), 2),
                  (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')


plt.tight_layout(pad=4)
fig.show()
```
